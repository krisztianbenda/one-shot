```python
from PIL import Image
import sys
import numpy as np
from scipy.misc import imread
import cPickle as pickle
import os
import matplotlib.pyplot as plt
import argparse
from PIL import Image
import csv
```

## Fontosabb valtozok

    @data_path: Az osszerendezett testverparok mappajanak helye

    @target_size: A program negyzetes meretre vagja a kepeket. A negyzet oldalanak hosszat tudjuk target_size-zal beallitani

    @base_size: Az eredeti kepek merete. Mivel a SiblingsDB-ben tobb meretu kep is szerepel, ezert a leggyakoribb meret. Ha csak lehet ne allitsuk at.

```python
data_path = "SiblingsDB/DBs/HQf_withPairs"
target_size = 105 
base_size = (4256,2832)
```

```python
def create_empty_image(path):
    img = Image.new('RGB', (target_size, target_size), (123, 101, 52))
    img.save(path)

def white_bg_square(img):
    # return a white-background-color image having the img in exact center
    layer = Image.new('RGB', base_size, (255,255,255))
    layer.paste(img, tuple(map(lambda x:(x[0]-x[1])/2, zip(base_size, img.size))))
    return layer

def cut_edges(img):
    return img.crop(((img.size[0]-img.size[1])/2, 0, (img.size[0]-img.size[1])/2 + img.size[1], img.size[1]))
    
def resize_image(image, convert_bw, signal):
    outfile = os.path.splitext(image)[0] + signal +".jpeg"
    img = Image.open(image)
    if img.size != base_size:
        img = white_bg_square(img)
    temp_width = int(float(target_size) / float(img.size[1]) * float(img.size[0]))
    img = img.resize((temp_width, target_size), Image.ANTIALIAS)
    img = cut_edges(img)
    if convert_bw:
        img = img.convert('1')
    img.save(outfile, "jpeg")
    print('SAVED: ' + outfile)
    
def delete_images(signal = 'new'):
    # Delete all signaled files
    for staff in os.listdir(data_path):
        staff_path = os.path.join(data_path,staff)
        if ".DS_Store" in staff_path:
            os.remove(staff_path)
        else:
            for filename in os.listdir(staff_path):
                if ".DS_Store" in filename or signal in filename:
                    os.remove(staff_path + '/' + filename)
                    
def get_coordinates(from_csv, point_number):
    with open(from_csv, 'rb') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',')
        for row in spamreader:
            if row[0] == str(point_number):
                return int(float(row[1])), int(float(row[2]))
        return "Error, there is not any row int csv with that point_number"
        
def crop_image(image, x, y, width, height):
    return image.crop((x-int(width/2),y-int(height/2),x+int(width/2),y+int(height/2)))

def create_cropped_images(path, signal, point_number, default_size, black_white=False):
    for sibling in os.listdir(path):
        sibling_path = os.path.join(path,sibling)
        print('Processing family ' + sibling_path[-4:].split('/')[-1] +' with signal: ' + signal)
        if ".DS_Store" not in sibling_path and int(sibling):
            for person in os.listdir(sibling_path):
                if ".csv" in person:
                    csv_path = os.path.join(sibling_path, person)
                    x, y = get_coordinates(csv_path, point_number)
                    new_img = crop_image(Image.open(csv_path[:-4] + '.jpg'),
                                        x, y, default_size, default_size)
                    new_img.thumbnail((target_size, target_size))
                    if black_white:
                        new_img = new_img.convert('1')
                    new_img.save(csv_path[:-4] + signal + '.jpeg', 'jpeg')
```

## Kepek kivagasa

Az implementalt fuggvenyekkel tudjuk elkesziteni a kivagott kepeket.
A create_cropped_images() fuggvenny hajtja vegre a feladatokat.
Parameterei:

    @path: A testverparok kepeinek helye
    @signal: Egy karakteres jeloles az uj kepeknevehez. 
    @point_number: Annak a pontnak a szama amelyet korul szeretnenk vagni. 
        A pontok elhelyezkedeset a SiblingsDB - content description 2.0.pdf SiblingsDB-vel letoltottfajl tartalmazza
    @default_size: az eredeti kepen vett kivagas merete
    @black_white: True eseten fekete feher kepeket keszit

```python
create_cropped_images(data_path, 'eye', 32, 305, black_white=True)  
create_cropped_images(data_path, 'nose', 68, 420, black_white=True)  
create_cropped_images(data_path, 'mouth', 67, 420, black_white=True) 
```

```python
def load_images(path, signal, from_=0, to=200):
    X = []
    y = []
    cat_dict = {}
    family_dict = {}
    curr_y = 0
    for family in os.listdir(path):
        family_dict[family] = [curr_y,None]
        siblings_path = os.path.join(path,family)
        category_images=[]
        if ".DS_Store" not in siblings_path and int(family) >= from_ and int(family) <= to:
            for person in os.listdir(siblings_path):
                if ('.jpg' in person):
                    resize_image(siblings_path + '/' + person,  True, signal)
            image_num = 0
            for filename in os.listdir(siblings_path):
                if (signal in filename and image_num < 2):
                    image_path = os.path.join(siblings_path, filename)
                    image = imread(image_path)
                    category_images.append(image)
                    y.append(curr_y)
                    image_num +=1
            X.append(np.stack(category_images))
            curr_y += 1
            family_dict[family][1] = curr_y - 1
            
    y = np.vstack(y)
    X = np.stack(X)
    print('STORED ' + signal + ' imgs in '+ path + ': ' + str(curr_y) + '\n')
    return X,y,family_dict
```

## Kepobjektumok elmentese

A kovetkezo utasitasokkal elmentesekre kerulnek a megadott jelu kepek.

A funkciot ellato fuggveny a load_images(), melyenk parameterei a kovetkezok:

        @path: a testverparokat tartalmazo mappa
        @signal: a megadott jelzesu kepek kerulnek mentesre
        @from_: a csaladokat szamok azonositjak. Ettol a csaladtol kezdodoen tortenik a feldolgozas
        @to: eddig a csaladig tortenik a feldolgozas
Ha eddig meg nem latott jelet adunk meg parameterkent, akkor ujat fog letrehozni, hasonloan a teljes arcos kepekhez.

```python
signals = ['full', 'eye', 'mouth', 'nose']

for i in range(0, len(signals)):
    X,y,c=load_images(data_path, signals[i], from_=0, to=172)
    with open("train_sib_" + signals[i]+ ".pickle", "wb") as f:
        pickle.dump((X,c),f, protocol=2)


    X,y,c=load_images(data_path, signals[i], from_=173, to=300)
    with open("val_sib_" + signals[i]+ ".pickle", "wb") as f:
        pickle.dump((X,c),f)
```
